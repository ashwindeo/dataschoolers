```python
sdf1 = raw_sdf.copy(deep=True)
# Drop Columns
sdf1.drop(u'Unnamed: 0', axis=1, inplace=True)
sdf1.drop(u'State Name [Public School] 2013-14', axis=1, inplace=True)
sdf1.drop(u'School Name [Public School] 2013-14', axis=1, inplace=True)
sdf1.drop(u'ANSI/FIPS State Code [Public School] Latest available year', axis=1, inplace=True)
sdf1.drop(u'Location Address [Public School] 2013-14', axis=1, inplace=True)
sdf1.drop(u'Location State Abbr [Public School] 2013-14', axis=1, inplace=True)
sdf1.drop(u'Location ZIP [Public School] 2013-14', axis=1, inplace=True)
sdf1.drop(u'Location ZIP4 [Public School] 2013-14', axis=1, inplace=True)
sdf1.drop(u'Mailing Address [Public School] 2013-14', axis=1, inplace=True)
sdf1.drop(u'Mailing State Abbr [Public School] 2013-14', axis=1, inplace=True)
sdf1.drop(u'Mailing City [Public School] 2013-14', axis=1, inplace=True)
sdf1.drop(u'Mailing ZIP [Public School] 2013-14', axis=1, inplace=True)
sdf1.drop(u'Mailing ZIP4 [Public School] 2013-14', axis=1, inplace=True)
sdf1.drop(u'Phone Number [Public School] 2013-14', axis=1, inplace=True)


sdf1.shape
```

```python
# Rename Columns
sdf1.rename(columns={'State Name [Public School] Latest available year' :'State Name'     }, inplace=True)
sdf1.rename(columns={'State Abbr [Public School] Latest available year' :'State'          }, inplace=True)

sdf1.rename(columns={'School ID - NCES Assigned [Public School] Latest available year' :'NCES SchoolID'  }, inplace=True)
sdf1.rename(columns={'State School ID [Public School] 2013-14'          :'State SchoolID' }, inplace=True)

sdf1.rename(columns={'Agency Name [Public School] 2013-14'              :'Agency Name'    }, inplace=True)
sdf1.rename(columns={'Agency ID - NCES Assigned [Public School] Latest available year' :'NCES AgencyID' }, inplace=True)
sdf1.rename(columns={'State Agency ID [Public School] 2013-14'          :'State AgencyID' }, inplace=True)

sdf1.rename(columns={'County Name [Public School] 2013-14'              :'County Name'    }, inplace=True)
sdf1.rename(columns={'County Number [Public School] 2013-14'            :'County Number'  }, inplace=True)

sdf1.rename(columns={'Years School Reported Data [Public School] Latest available year' :'Data Available'}, inplace=True)
sdf1.rename(columns={'Years School Did Not Report Data [Public School] Latest available year' :'Data NA'}, inplace=True)

sdf1.rename(columns={'Location City [Public School] 2013-14' :'City'}, inplace=True)


sdf1[:10]
```
